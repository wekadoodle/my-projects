```python
#Problem: POS Machine
#My Solution
class POSMachine:
    def __init__(self):
        self.__total = []
        self.__counter = 1
        self.__df = 'Please set a price list first.'
        self.__invoice1 = 'Sales Invoice\nID\tProduct\tPrice\tQuantity\tTotal'
        self.__all = {}
        self.start()
        
    def start(self): 
        self.printWelcomeMessage()
        self.printAllOptions()
        inp = input().strip()
        while inp != 'Y':
            if inp == '1':
                self.setPriceList()
            elif inp == '2':
                self.printPriceList()                
            elif inp == '3':
                self.makeNewTransaction()
            elif inp == '4':
                self.createSalesReport()
            elif inp == '5':
                inp = input('Are you sure you want to exit the program [Y]? Enter any key to go back.').upper().strip()
                if inp == 'Y':
                    break
                else:
                    self.printAllOptions()
                    inp = input().strip()
                    continue
            else:
                print('Please enter a valid option.')
            self.printAllOptions()
            inp = input().strip()
                
    def printWelcomeMessage(self): 
        
        print('Welcome to PP-POS System')
            
    def printAllOptions(self):
        
        print("What do you want to do today? (input the number)")
        print("1 - Set the price list")
        print("2 - Show the price list")
        print("3 - Make a new transaction")
        print("4 - Create a Sales Report")
        print("5 - Close the POS machine")
    
    def setPriceList(self):
        
        filepath = input('Enter the path to the input file:')
        while filepath[-4:] != '.txt' and filepath[-4:] != '.csv':
            print('Invalid file type. Supported types are .txt and .csv.')
            filepath = input('Enter the path to the input file:')
        if filepath[-4:] == '.txt':
            sep_perator = "\t"
        else:
            sep_perator = ','
        try:
            self.__df = pd.read_csv(filepath, sep = sep_perator).set_index('ID')
            sdf = pd.read_csv(filepath, sep = sep_perator).set_index('Product')
            if self.__all != {}:
                for key in self.__all:
                    self.__all[key][0] = sdf.loc[key]['Price']     
                    
        except:
            print('File not found.')
            
    def printPriceList(self):
        
        print(self.__df)
    
    def makeNewTransaction(self):
        
        self.__total.clear()
        self.__invoice = 'Sales Invoice\nID\tProduct\tPrice\tQuantity\tTotal'
        productID = ' ' #placeholder
        
        while type(self.__df) == type(pd.DataFrame()) and productID != '': #checks if price list was set
            productID = input('Product ID: ').strip() #initial input
            
            while productID != '':
                while productID.isnumeric() == False and productID != '':
                    print('Please enter a valid input.')
                    productID = input('Product ID: ').strip()
                    
                productID = int(productID) 
                    
                if productID not in self.__df.index.tolist():
                    print('Product not found.')
                    productID = input('Product ID: ').strip()
                            
                else:
                    quantity = input('Quantity: ').strip()                    
                    while quantity != '':
                        while quantity.isnumeric() == False or int(quantity) <= 0:
                            print('Please enter a valid input.')
                            quantity = input('Quantity: ').strip()                        
                            if quantity == '':
                                break
                        if quantity != '':
                            quantity = int(quantity)
                            self.addProductToTransaction(productID,quantity)
                            break
                    if quantity != '':
                        productID = input('Product ID: ').strip()
                    else:
                        productID = ''
                        break
                
        if len(self.__total) > 0:
            self.createSalesInvoice()
        else:
            print('Please set a price list first.')
    
    def addProductToTransaction(self, productID, quantity):
        #not in self.__all
        if self.__df.loc[productID]['Product'] not in self.__all:
            self.__all[self.__df.loc[productID]['Product']] = [self.__df.loc[productID]['Price'], quantity, self.__df.loc[productID]['Price'] * quantity]
        #in self.__all
        else:
            self.__all[self.__df.loc[productID]['Product']][0] = self.__df.loc[productID]['Price']
            self.__all[self.__df.loc[productID]['Product']][1] += quantity
            self.__all[self.__df.loc[productID]['Product']][2] += self.__df.loc[productID]['Price'] * quantity
        
        self.__total.append(self.__df.loc[productID]['Price'] * quantity)
        self.__invoice = self.__invoice + '\n{id}\t{product}\t{price}\t{quantity}\t{total}'.format(id = productID,
                                                                                 product = self.__df.loc[productID]['Product'],
                                                                                 price = self.__df.loc[productID]['Price'],
                                                                                 quantity = quantity,
                                                                                 total = self.__df.loc[productID]['Price'] * quantity)
    
    def createSalesInvoice(self):
        print(self.__invoice)
        print('\t\t\tGrand Total\t{}'.format(sum(self.__total)))
        with open('invoice{}.txt'.format(self.__counter), 'w') as f:
            f.write(self.__invoice)
            f.write('\n\t\t\tGrand Total\t{}'.format(sum(self.__total)))
        self.__counter +=1

    def createSalesReport(self):
        dfSalesReport = pd.DataFrame.from_dict(self.__all, orient = 'index', columns = ['Price', 'Total Number Sold', 'Total Sales']) #https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.from_dict.html
        print(dfSalesReport)
        
        
```
